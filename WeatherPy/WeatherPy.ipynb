```python
# Import Dependencies 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import scipy.stats as st
from pprint import pprint

# Import API key
from config import api_key

# Import Citipy to determine city by latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "./Output/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# Create empty lists for Latitude/Longitude and Cities
latLng = []
cities = []

# Create a set of random Latitudes and Longitudes
lat = np.random.uniform(low=-90.000, high=90.000, size=1500)
lng = np.random.uniform(low=-180.000, high=180.000, size=1500)
latLng = zip(lat, lng)

# Match cities with their corresponding Latitude and Longitude
for x in latLng:
    city = citipy.nearest_city(x[0],x[1]).city_name

    if city not in cities:
        cities.append(city)

len(cities)
```

```python
print(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
print(query_url)
```

```python
# Create empty lists for data information
weather_dict = {
    "City" : cities,
    "Country" : None,
    "Latitude" : None,
    "Longitude" : None,
    "Temperature (F)" : None,
    "Humidity (%)" : None,
    "Cloudiness (%)" : None,
    "Wind Speed (MPH)" : None
}

weather_df = pd.DataFrame(weather_dict).set_index("City")
```

```python
# Create a for loop 
x=1
for i, row in weather_df.iterrows():
    city = i

    try:
        # Make the request and print which place it's looking up
        response = requests.get(query_url + city).json()
        print(f"Retrieving data for city number {x}, {i}")
        x+=1
        try:
            # If there was a result, add it to the df
            if response and len(response):
                weather_df.loc[i, "Country"] = response["sys"]["country"]
                weather_df.loc[i, "Latitude"] = response["coord"]["lat"]
                weather_df.loc[i, "Longitude"] = response["coord"]["lon"]
                weather_df.loc[i, "Temperature (F)"] = response["main"]["temp_max"]
                weather_df.loc[i, "Humidity (%)"] = response["main"]["humidity"]
                weather_df.loc[i, "Cloudiness (%)"] = response["clouds"]["all"]
                weather_df.loc[i, "Wind Speed (MPH)"] = response["wind"]["speed"]

        except KeyError:
            print("Missing values....skipping")


    except KeyError:
        print("index can't be found....skipping")
```

```python
# Drop empty data points
weather_df = weather_df.dropna()
weather_df
```

```python
# Convert data types from object to float for later use in calculations
weather_df['Latitude'] = weather_df['Latitude'].astype(float)
weather_df['Longitude'] = weather_df['Longitude'].astype(float)
weather_df['Temperature (F)'] = weather_df['Temperature (F)'].astype(float)
weather_df['Humidity (%)'] = weather_df['Humidity (%)'].astype(float)
weather_df['Cloudiness (%)'] = weather_df['Cloudiness (%)'].astype(float)
weather_df['Wind Speed (MPH)'] = weather_df['Wind Speed (MPH)'].astype(float)
weather_df
```

```python
# Save Output file to folder
weather_df.to_csv(output_data_file)
```

```python
# Create a function runs a full Linear Regression given the variables
def LinearRegressionPlot(hemisphere, x_values, y_values, x_label, ylabel, title, savename):

    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    
    plt.annotate(line_eq, (0,40), fontsize=15,color="red")
    print(f"The r-squared is: {rvalue**2}")
    correlation = st.pearsonr(x_values,y_values)
    print(f"The correlation between both factors is {round(correlation[0],2)}")

    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    
    plt.savefig(f"./Images/{savename}.png")
        
    plt.grid()
    plt.show()
    
```

```python
# Run linear regression for Northern Hemisphere vs Temperature
hemisphere = north_df
x_values = north_df['Latitude']
y_values = north_df['Temperature (F)']
x_label = "Latitude"
y_label = "Temperature (F)"
title = "Northern Latitude vs Temperature (F)"
savename = "northern_temp"

LinearRegressionPlot(hemisphere, x_values, y_values, x_label, y_label, title, savename)
```
